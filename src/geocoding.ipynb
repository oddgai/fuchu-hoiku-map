```python
import requests
from urllib.parse import quote
import xml.etree.ElementTree as ET
import time
import pandas as pd
```

```python
def get_coordinate(row):
    point_name = "府中 " + row["施設名"]
    print(point_name)
    url = f"https://www.geocoding.jp/api/?q={quote(point_name)}"
    res = requests.get(url)
    time.sleep(10)
    if res.status_code == 200:
        xml_root = ET.fromstring(res.text.replace("&amp;", "＆"))
        return pd.Series([float(xml_root[2][0].text), float(xml_root[2][1].text)])
    else:
        return pd.Series([None, None])
```

```python
hoiku_data = pd.read_table("認可外.tsv")
```

```python
hoiku_data[["lat", "lon"]] = hoiku_data.apply(get_coordinate, axis=1)
```
